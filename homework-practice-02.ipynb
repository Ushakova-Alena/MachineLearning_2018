```python
def prod_non_zero_diag(x):
    prod = 1
    diag_index = 0
    index_max = len(x[0]) - 1
    for line in x:
        if diag_index > index_max:
            break
        elem = line[diag_index]
        if elem != 0:
            prod *= elem
        diag_index += 1
    return prod


def are_multisets_equal(x):
    return sorted(x[0]) == sorted(x[1])

def max_after_zero(x):
    max_elem = None
    for i in range(len(x)):
        if i + 1 > len(x) - 1:
            break
        if x[i] == 0 and (max_elem == None or x[i+1] > max_elem):
            max_elem = x[i+1]
    return max_elem


def convert_image(x):
    img_new = []
    for i in range(len(x[0])):
        img_new.append([])
        for j in range(len(x[0][i])):
            img_new[i].append(sum([a*b for a,b in zip(x[0][i][j],x[1])]))
    return img_new


def run_length_encoding(x):
    elements = []
    counters = []
    times = 0
    prev_elem = None
    for elem in x:
        if prev_elem != elem:
            elements.append(elem)
            if times != 0:
                counters.append(times)
                times = 0
        times += 1
        prev_elem = elem
    counters.append(times)
    return (elements, counters)


def pairwise_distance(x):
    from math import sqrt
    dist = []
    for elem_x in x[0]:
        new_row = []
        for elem_y in x[1]:
            new_row.append(sqrt(sum((a-b)*(a-b) for a,b in zip(elem_x,elem_y))))
        dist.append(new_row)
    return dist
```

```python
import numpy as np


def prod_non_zero_diag_vect(x):
    return np.prod(x.diagonal()[np.nonzero(x.diagonal())])


def are_multisets_equal_vect(x):
    return np.array_equal(np.sort(x[0]), np.sort(x[1]))


def max_after_zero_vect(x):
    indices = np.where(x == 0)[0] + 1
    return max(np.take(x, indices, mode="clip"))


def convert_image_vect(x):
    return np.dot(x[0], x[1])


def run_length_encoding_vect(x):
    y = np.array(x[1:] != x[:-1])
    i = np.append(np.where(y), len(x) - 1)
    z = np.diff(np.append(-1, i))
    return x[i], z


def pairwise_distance_vect(x):
    diff = x[0][:,None,:] - x[1]
    return np.sqrt(np.einsum('abc,abc->ab',diff,diff)) 
```

```python
def pairplot(df, target):
    ncol, nrow = 7, df.shape[1] // 7 + (df.shape[1] % 7 > 0)
    plt.figure(figsize=(ncol * 4, nrow * 4))

    for i, feature in enumerate(df.columns):
        plt.subplot(nrow, ncol, i + 1)
        plt.scatter(df[feature], target, s=10, marker='o', alpha=.6)
        plt.xlabel(feature)
        if i % ncol == 0:
            plt.ylabel('target')
```
